# Import & Path Setup

```python
from pathlib import Path
import os
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt

BASE_DIR = Path.cwd().parents[0]
HR_DIR = BASE_DIR / "data" / "DIV2K" / "HR"
HR_PATCH_DIR = BASE_DIR / "data" / "DIV2K" / "HR_patches"
LR_PATCH_DIR = BASE_DIR / "data" / "DIV2K" / "LR_patches"

print("HR_DIR:", HR_DIR)
print("Jumlah HR:", len(glob(str(HR_DIR/"*.png"))))
```

# Tampilkan Contoh Gambar HR Asli

```python
img_path = glob(str(HR_DIR/"*.png"))[0]
img = Image.open(img_path)

plt.figure()
plt.imshow(img)
plt.title("Contoh Gambar HR Asli")
plt.axis("off")
```

# Proses Crop HR  Patch + LR

```python
sample_hr = img.crop((0, 0, 128, 128))
sample_lr = sample_hr.resize((64, 64), Image.BICUBIC)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(sample_hr)
plt.title("Patch HR 128x128")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sample_lr)
plt.title("Patch LR 64x64 (Bicubic)")
plt.axis("off")
```

# Statistik Jumlah Patch

```python
print("Jumlah HR patches:", len(glob(str(HR_PATCH_DIR/"*_HR.png"))))
print("Jumlah LR patches:", len(glob(str(LR_PATCH_DIR/"*_LR.png"))))
```
