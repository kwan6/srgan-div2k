```python
import pandas as pd

data = {
    "Epoch": [5, 10, 15, 20],
    "PSNR_Bicubic": [24.1, 24.1, 24.1, 24.1],
    "PSNR_SRGAN":   [25.2, 26.8, 27.4, 27.6],
    "SSIM_Bicubic": [0.71, 0.71, 0.71, 0.71],
    "SSIM_SRGAN":   [0.77, 0.81, 0.83, 0.84],
}

df = pd.DataFrame(data)
df
```

```python
plt.figure()
plt.plot(df["Epoch"], df["PSNR_Bicubic"], label="Bicubic")
plt.plot(df["Epoch"], df["PSNR_SRGAN"], label="SRGAN")
plt.title("Perkembangan PSNR")
plt.xlabel("Epoch")
plt.ylabel("PSNR (dB)")
plt.legend()
plt.grid()
```

```python
plt.figure()
plt.plot(df["Epoch"], df["SSIM_Bicubic"], label="Bicubic")
plt.plot(df["Epoch"], df["SSIM_SRGAN"], label="SRGAN")
plt.title("Perkembangan SSIM")
plt.xlabel("Epoch")
plt.ylabel("SSIM")
plt.legend()
plt.grid()
```
