```python
from PIL import Image
from glob import glob
import matplotlib.pyplot as plt
from pathlib import Path

BASE_DIR = Path.cwd().parents[0]
CUSTOM_DIR = BASE_DIR / "outputs" / "custom_results"

paths = sorted(glob(str(CUSTOM_DIR/"*.png")))
paths
```

```python
orig = Image.open(paths[2])
bic = Image.open(paths[1])
sr  = Image.open(paths[0])

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(orig)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(bic)
plt.title("Bicubic")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(sr)
plt.title("SRGAN")
plt.axis("off")
```
